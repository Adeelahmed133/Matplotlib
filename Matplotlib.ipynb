### Plotting x and y points:-

The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

If we need to plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.

```python
# Draw a line in a diagram from position (1,3) to position (8, 10)
apoint = np.array([1, 8])
bpoint = np.array([3, 10])
plt.plot(apoint, bpoint)
plt.show()
```

### Plotting Without Line:-

To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.

```python
cpoint = np.array([1, 8])
dpoint = np.array([3, 10])
plt.plot(cpoint, dpoint, 'o') # Graph show in o (O) points.
plt.show()
```

### Set Font Properties for Title and Labels:-

You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.

```python
font1 = {'family' : 'serif', 'color' : 'blue'}
font2 = {'family' : 'serif',  'color' : 'red', 'size' : 20}
font3 = {'family' : 'serif', 'color' : 'green', 'size' : 10}
plt.title('Sports Watch Data', fontdict = font1)
plt.xlabel('Average Pulse', fontdict = font2)
plt.ylabel('Calorie Burnage', fontdict = font3)
plt.show()
```

### Position the Title:-

You can use the loc parameter in title() to position the title.

Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

```python
g1 = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
h1 = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.plot(g1, h1)
plt.show()
```

### Grid Lines:-

Add Grid Lines to a Plot With Pyplot, you can use the grid() function to add grid lines to the plot.

```python
i1 = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
j1 = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.plot(i1, j1)
plt.grid()
plt.show()
```

### Multiple Plots:-

Display Multiple Plots With the subplot() function you can draw multiple plots in one figure.

```python
# Make 2 plots
#plot 1:
m = np.array([0, 1, 2, 3])
n = np.array([3, 8, 1, 10])
plt.subplot(1, 2, 1)
plt.plot(m,n)
#plot 2:
m = np.array([0, 1, 2, 3])
n = np.array([10, 20, 30, 40])
plt.subplot(1, 2, 2)
plt.plot(m,n)
plt.show()
```

You can add a title to each plot with the title() function.

### Super Title:-

You can add a title to the entire figure with the suptitle() function.

```python
#plot 1:
x3 = np.array([0, 1, 2, 3])
y3 = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x3,y3)
plt.title("SALES")

#plot 2:
x3 = np.array([0, 1, 2, 3])
y3 = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x3,y3)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```
